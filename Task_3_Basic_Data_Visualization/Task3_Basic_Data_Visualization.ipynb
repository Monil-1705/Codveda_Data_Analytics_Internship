```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the Dataset

```python
df = pd.read_csv("../Datasets/1) iris.csv")
```

```python
df
```

## Bar Plots

```python
sns.barplot(data=df, x="species", hue="species", y="sepal_length", palette='pastel', legend=False)
plt.title("Bar plot of Sepal Length vs Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length")
plt.grid(axis="y")
plt.show()
```

```python
sns.barplot(data=df, x="species", hue="species", y="sepal_width", palette='pastel', legend=False)
plt.title("Bar plot of Sepal Width vs Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Width")
plt.grid(axis="y")
plt.show()
```

```python
sns.barplot(data=df, x="species", hue="species", y="petal_length", palette='pastel', legend=False)
plt.title("Bar plot of Petal Length vs Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Length")
plt.grid(axis="y")
plt.show()
```

```python
sns.barplot(data=df, x="species", hue="species", y="petal_width", palette='pastel', legend=False)
plt.title("Bar plot of Petal Width vs Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Width")
plt.grid(axis="y")
plt.show()
```

## Line Charts

```python
species_sl_mean = df.groupby('species')['sepal_length'].mean().reset_index()
sns.lineplot(data=df, x=species_sl_mean['species'], y=species_sl_mean['sepal_length'], marker='o', linestyle='-', color="blue")
plt.title("Line plot of Mean Sepal Length vs Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length")
plt.grid(axis="y")
plt.show()
```

```python
species_sw_mean = df.groupby('species')['sepal_width'].mean().reset_index()
sns.lineplot(data=df, x=species_sw_mean['species'], y=species_sw_mean['sepal_width'], marker='o', linestyle='-', color="yellow")
plt.title("Line plot of Mean Sepal Width vs Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Width")
plt.grid(axis="y")
plt.show()
```

```python
species_pl_mean = df.groupby('species')['petal_length'].mean().reset_index()
sns.lineplot(data=df, x=species_pl_mean['species'], y=species_pl_mean['petal_length'], marker='o', linestyle='-', color="green")
plt.title("Line plot of Mean Petal Length vs Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Length")
plt.grid(axis="y")
plt.show()
```

```python
species_pw_mean = df.groupby('species')['petal_width'].mean().reset_index()
sns.lineplot(data=df, x=species_pw_mean['species'], y=species_pw_mean['petal_width'], marker='o', linestyle='-', color="pink")
plt.title("Line plot of Mean Petal Width vs Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Width")
plt.grid(axis="y")
plt.show()
```
