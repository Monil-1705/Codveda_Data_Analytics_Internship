```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the dataset

```python
df = pd.read_csv("../Datasets/1) iris.csv")
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
mean_list = []
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        mean_list.append(df[col].mean())
        print(f"Mean of {col} is : {df[col].mean()}")
```

```python
mean_list
```

```python
median_list = []
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        median_list.append(df[col].median())
        print(f"Median of {col} is : {df[col].median()}")
```

```python
median_list
```

```python
mode_list = []
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        mode_list.append(df[col].mode())
        print(f"Mode of {col} is : {df[col].mode()[0]}")
```

```python
stdev_list = []
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        stdev_list.append(df[col].std())
        print(f"Standard Deviation of {col} is : {df[col].std()}")
```

# Visualizations : 

### Histograms

```python
sns.histplot(data=df, x='sepal_length', color="blue", bins=10)
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(data=df, x="sepal_width", color='green', bins=10)
plt.title("Histogram of Sepal Width")
plt.xlabel("Sepal Width")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(data=df, x='petal_length', color='yellow', bins=10)
plt.title("Histogram of Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()
```

### Box Plots

```python
sns.boxplot(data=df, y='sepal_length', color="blue")
plt.title("Box plot of Sepal Length")
plt.show()
```
