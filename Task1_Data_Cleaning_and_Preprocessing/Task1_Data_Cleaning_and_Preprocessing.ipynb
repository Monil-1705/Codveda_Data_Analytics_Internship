## Handling Categorical Columns

### International plan

```python
df.loc[df['International plan'] == "Yes", 'International plan'] = "1"
```

```python
df.loc[df['International plan'] == "No", 'International plan'] = "0"
```

```python
df['International plan']
```

```python
df['International plan'] = df['International plan'].astype('int64')
```

```python
df['International plan'].dtype
```

### Voice mail plan

```python
df.loc[df['Voice mail plan'] == "Yes", 'Voice mail plan'] = "1"
```

```python
df.loc[df['Voice mail plan'] == "No", 'Voice mail plan'] = "0"
```

```python
df['Voice mail plan']
```

```python
df['Voice mail plan'] = df['Voice mail plan'].astype('int64')
```

```python
df['Voice mail plan'].dtype
```

### Churn

```python
df['Churn'] = df['Churn'].astype("O")
```

```python
df.loc[df['Churn'] == "True", 'Churn'] = "1"
```

```python
df.loc[df['Churn'] == "False", 'Churn'] = "0"
```

```python
df['Churn'] = df['Churn'].astype('int64')
```
